```python
import pandas as pd
df=pd.read_csv("/content/sample_data/california_housing_train.csv")
df.head()
print(df.isnull().sum())
print(df.duplicated().sum())
print(df.info())
import matplotlib.pyplot as plt
df["median_house_value"].hist()
plt.title("Median House Value Distribution")
plt.show()
df.groupby("housing_median_age")["median_house_value"].mean().plot(kind="bar")
plt.title("Average Median House Value by Housing Median Age")
df.plot.scatter(x="median_income",y="median_house_value")
plt.title("median income vs house value")
plt.show()
df.plot.scatter(x="total_rooms", y="median_house_value")
plt.title("Total Rooms vs Median House Value")
plt.show()
```
